```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glas of milk',
      'the glass of juice',
      'i am a good boy',
      'i am a bad developer',
      'understand the meaning of words',
      'your videos are good ']
```

```python
sent
```

```python
voc_size= 10000
```

```python
one_hot_repr=[one_hot(words,voc_size) for words in sent] # this will give us the one hot representation of the words in the sentences
```

```python
one_hot_repr
```

```python
## word emding layar 
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
# pad_sequences is used to make the length of all the sentences equal by adding 0's in the beginning of the sentences which are smaller in length than the maximum length sentence.
# otherwise the model will not be able to process the sentences of different lengths.
```

```python
import numpy as np
```

```python
set_length =8 
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=set_length)
print(embedded_docs )
# here we have used padding='pre' which means that the 0's will be added in the beginning of the sentences which are smaller in length than the maximum length sentence.
```

```python
# feature representation of the sentences
dim= 10 
```

```python
model= Sequential()
model.add(Embedding(voc_size,dim,input_length=set_length))# here we have used voc_size as the input dimension which is the size of the vocabulary, dim as the output dimension which is the size of the embedding vector and set_length as the input length which is the length of the sentences. 
model.compile('adam','mse') # here we have used adam as the optimizer and mse as the loss function.
```

```python
model.predict(embedded_docs) # this will give us the feature representation of the sentences in the form of a 3D array. The shape of the array will be (number of sentences, length of sentences, size of embedding vector).
```

```python
model.predict([0,0,0,0,0,0,0,0]) # this will give us the feature representation of the sentences in the form of a 3D array. The shape of the array will be (number of sentences, length of sentences, size of embedding vector).
```
